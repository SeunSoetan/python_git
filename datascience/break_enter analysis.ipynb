```python
sub_breakenter_df = breakenter_dropna[['X', 'Y', 'occurrencedate','occurrenceyear', 'occurrencemonth', 'occurrenceday','occurrencedayofyear','occurrencedayofweek','occurrencehour','Neighbourhood','offence']]
```

```python
grouped_neighbourhood_df = sub_breakenter_df[['Neighbourhood', 'offence']].groupby('Neighbourhood', as_index=False).count()
```

```python
print(grouped_neighbourhood_df)
```

```python
grouped_neighbourhood_df.offence.sum()
```

```python
fig, ax = plt.subplots(figsize=(28,20))
ax.bar(grouped_neighbourhood_df.Neighbourhood, grouped_neighbourhood_df.offence)
```

```python
grouped_neighbourhood_df_400more = grouped_neighbourhood_df[grouped_neighbourhood_df.offence >= 400]
```

```python
print(grouped_neighbourhood_df_400more)
```

```python
grouped_neighbourhood_df_400more_sort = grouped_neighbourhood_df_400more.sort_values(['offence'], ascending=False)
```

```python
print(grouped_neighbourhood_df_400more_sort)
```

```python
fig, ax = plt.subplots(figsize=(28,20))
ax.bar(grouped_neighbourhood_df_400more_sort.Neighbourhood, grouped_neighbourhood_df_400more_sort.offence)
plt.xlabel('Neighbourhood')
plt.ylabel('Rate of Crime in Neighbourhood')
```

```python
grouped_neighbourhood_df_sort = grouped_neighbourhood_df.sort_values(['offence'], ascending=False)
```

```python
grouped_neighbourhood_df_sort.head()
```

```python
grouped_neighbourhood_df_sort.tail()
```

```python
fig = plt.subplots(figsize=(18,10))
plt.subplot(2,1,1)
plt.bar(grouped_neighbourhood_df_sort.Neighbourhood[:5], grouped_neighbourhood_df_sort.offence[:5])
plt.xlabel('Neighbourhood')
plt.ylabel('Rate of Crime in Neighbourhood')
plt.title('The Top 5 Neighbourhood with the Highest Crime')


plt.subplot(2,1,2)
plt.bar(grouped_neighbourhood_df_sort.Neighbourhood[-6:-1], grouped_neighbourhood_df_sort.offence[-6:-1])
plt.xlabel('Neighbourhood')
plt.ylabel('Rate of Crime in Neighbourhood')
plt.title('The Least 5 Neighbourhood with the Lowest Crime')
```
